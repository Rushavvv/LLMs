The pipeline() function:

It connects a model with its necessary preprocessing and postprocessing steps, allowing us to directly input any text and get an intelligible answer

```python
from transformers import pipeline
```

Using bert-base-uncased. This is only a pretrained encoder [base model].Not trained for any specific downstream task. This base model is without any task-specific head (It doesn’t know if you want to classify text, tag tokens, or predict next words.)

```python
classifier = pipeline("sentiment-analysis", model="bert-base-uncased")
classifier("I've been teaching NLP course to bachelor students. I do not know whether they are making sense of this course.")
```

What is Label_0?

Since we didn’t fine-tune it on sentiment data, the model hasn’t learned what “positive” or “negative” means. It still outputs a number for each label (LABEL_0, LABEL_1), but those numbers are meaningless until training aligns them with sentiment.

-

When we use a fine-tuned checkpoint like distilbert-base-uncased-finetuned-sst-2-english (Which is given to us by default): The classification layer has been trained on a dataset (SST-2) with labeled sentiment examples. The config file stores id2label = {0: 'NEGATIVE', 1: 'POSITIVE'}. Predictions are meaningful and labeled correctly.

```python
classifier = pipeline("sentiment-analysis")
classifier("I've been teaching NLP course to bachelor students. I do not know whether they are making sense of this course.")
```

```python
classifier(
    ["I've been waiting for a HuggingFace course my whole life.", "I hate this so much!"]
)
```

3-class (negative/neutral/positive):model="cardiffnlp/twitter-roberta-base-sentiment-late

```python
three_class_classification = pipeline("sentiment-analysis", model="cardiffnlp/twitter-roberta-base-sentiment-latest", device=0)
```

```python
three_class_classification("I've been teaching NLP course to bachelor students. I do not know whether they are making sense of this course.")
```

```python
three_class_classification("I've been teaching NLP course to bachelor students. Some of them has gained some knowledge.")
```

If we do not have to finetuned but want to classify from our targeted classes.

```python
classifier = pipeline("zero-shot-classification")
classifier(
    "This is a course about the Transformers library",
    candidate_labels=["education", "politics", "business"],
)
```
